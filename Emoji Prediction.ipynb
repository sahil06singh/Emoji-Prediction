```python
import emoji
import pandas as pd
import numpy as np
from tensorflow import keras
from keras.utils import to_categorical
import warnings
warnings.filterwarnings('ignore')
from keras.callbacks import ModelCheckpoint
from keras.callbacks import EarlyStopping
```

```python
train = pd.read_csv("dataset/train_emoji.csv" , header=None)
test = pd.read_csv("dataset/test_emoji.csv" , header=None )
```

```python
train.head()
```

```python
#creating emoji dictionary  (key-label , value->emoji name)
emoji_dictionary = {"0": "\u2764\uFE0F",    # :heart: prints a black instead of red heart depending on the font
                    "1": ":baseball:",
                    "2": ":beaming_face_with_smiling_eyes:",
                    "3": ":downcast_face_with_sweat:",
                    "4": ":fork_and_knife:",
                   }
```

```python
for value in emoji_dictionary.values():
    print(emoji.emojize(value))
```

```python
X_train = train[0]
X_test  = test[0]
Y_train = to_categorical(train[1])
Y_test = to_categorical(test[1])

```

```python
embeddings = {}

with open("glove.6B.50d.txt" , encoding='utf-8') as f:
    for line in f:
        line = line.split()
        word = line[0]
        coeff = np.asarray(line[1:] , dtype='float32')
        embeddings[word] = coeff
```

```python
def getOutputEmbeddings(X):
    
    embedding_matrix_output = np.zeros((X.shape[0],10,50))
    for i in range(X.shape[0]):
        X[i] = X[i].split()
        for j in range(len(X[i])):
            embedding_matrix_output[i][j] = embeddings[X[i][j].lower()]
            
    return embedding_matrix_output
```

```python
embeddingTrain = getOutputEmbeddings(X_train)
embeddingTest = getOutputEmbeddings(X_test)
```

```python
print(embeddingTrain.shape)
print(embeddingTest.shape)
```

```python
checkpoint =  keras.callbacks.ModelCheckpoint('bestEmoji.hdf5',  monitor='val_loss',mode='min',save_best_only=True)
modelLSTM  = keras.Sequential([
    keras.layers.LSTM(64 , input_shape = (10,50)),
    keras.layers.Dropout(0.35),
    keras.layers.Dense(5,activation = 'softmax')
])

modelLSTM.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
modelLSTM.fit(embeddingTrain,Y_train,validation_split=0.1,batch_size=65,epochs = 200,shuffle=True , callbacks = [checkpoint])
```

```python
modelLSTM.load_weights("bestEmoji.hdf5")
```

```python
modelLSTM.evaluate(embeddingTest,Y_test)
```

```python
checkpoint =  keras.callbacks.ModelCheckpoint('bestStacked.hdf5',  monitor='val_loss',mode='min',save_best_only=True)
modelStacked  = keras.Sequential([
    keras.layers.LSTM(64 , input_shape = (10,50)  ,return_sequences = True),
    keras.layers.Dropout(0.4),
    keras.layers.LSTM(64 , return_sequences = False),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(5,activation = 'softmax')
])

modelStacked.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
modelStacked.fit(embeddingTrain,Y_train,batch_size=32,epochs=100,shuffle=True,validation_split=0.1,callbacks = [checkpoint])
```

```python
modelStacked.load_weights('bestStacked.hdf5')
```

```python
modelStacked.evaluate(embeddingTest,Y_test)
```
